```python
quiz_questions = [
    {'question': f'1. What is {title}?', 'options': ['A) Option A', 'B) Option B', 'C) Option C', 'D) Option D'], 'answer': 'B'},
    {'question': '2. Question 2?', 'options': ['A) Answer A', 'B) Answer B', 'C) Answer C', 'D) Answer D'], 'answer': 'B'},
    {'question': '3. Question 3?', 'options': ['A) Answer A', 'B) Answer B', 'C) Answer C', 'D) Answer D'], 'answer': 'C'},
    {'question': '4. Question 4?', 'options': ['A) Answer A', 'B) Answer B', 'C) Answer C', 'D) Answer D'], 'answer': 'A'},
    {'question': '5. Question 5?', 'options': ['A) Answer A', 'B) Answer B', 'C) Answer C', 'D) Answer D'], 'answer': 'D'}
]

print('='*50)
print(f'  Day {day_num} Quiz: {title}')
print('='*50 + '\n')

score = 0
for q in quiz_questions:
    print(q['question'])
    for opt in q['options']:
        print(f'  {opt}')
    answer = input('Your answer: ').strip().upper()
    if answer == q['answer']:
        print('✓ Correct!\n')
        score += 1
    else:
        print(f"✗ Incorrect. The correct answer was {q['answer']}.\n")

print('='*50)
print(f'  Final Score: {score}/{len(quiz_questions)} ({score/len(quiz_questions)*100:.0f}%)')
print('='*50)
```
