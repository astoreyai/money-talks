### 1. Sole Proprietorship

**Structure:**
- No separate legal entity
- You ARE the business
- Report on Schedule C

**Advantages:**
```
✓ Zero setup cost
✓ Minimal paperwork
✓ Complete control
✓ All profits to owner
✓ Easy tax filing
```

**Disadvantages:**
```
✗ NO liability protection
✗ Personal assets at risk
✗ Harder to raise capital
✗ No separation from personal
```

**Best For:**
- New traders testing TTS
- Small trading operations (<$50K)
- Those wanting simplicity

### 2. Limited Liability Company (LLC)

**Structure:**
- Separate legal entity
- Flexible tax treatment
- State-level formation

**Tax Options:**
```
LLC can be taxed as:
├─ Disregarded Entity (default single-member)
│  └─ Like sole prop for taxes
├─ Partnership (default multi-member)
│  └─ K-1 distributions
├─ S-Corporation (by election)
│  └─ W-2 + distributions
└─ C-Corporation (by election)
   └─ Corporate taxation
```

**Advantages:**
```
✓ Liability protection
✓ Flexible taxation
✓ Simple to maintain
✓ Professional appearance
✓ Asset protection
```

**Disadvantages:**
```
✗ State filing fees
✗ Annual reports required
✗ More paperwork than sole prop
✗ Potential franchise taxes
```

**Setup Process:**
1. Choose state (usually your home state)
2. File Articles of Organization
3. Create Operating Agreement
4. Obtain EIN from IRS
5. Open business bank account

**Cost Example:**
```
Formation Filing:        $100-$500 (varies by state)
Operating Agreement:     $0-$500
Registered Agent:        $50-$300/year
Annual Report:           $0-$500/year
Total First Year:        $150-$1,800
```

### 3. S-Corporation

**Structure:**
- IRS tax election (not entity type)
- Usually LLC electing S-Corp status
- Pass-through taxation
- Must pay reasonable salary

**Key Feature: SE Tax Savings**
```
Without S-Corp:                    With S-Corp:
Trading profit: $150K              Salary (W-2):      $60K
SE tax: $0                         Distributions:     $90K
(Trading not subject to SE tax)    
                                   SE tax on salary:  $9,180
                                   No SE on dist:     $0
                                   
Note: For traders, S-Corp primarily useful for 
      other business income or consulting fees
```

**Advantages:**
```
✓ SE tax savings (on non-trading income)
✓ Liability protection
✓ Pass-through taxation
✓ Fringe benefit options
✓ Potential QBI deduction
```

**Disadvantages:**
```
✗ Must run payroll
✗ Reasonable compensation required
✗ More accounting costs
✗ Shareholder restrictions
✗ State filing requirements
```

**Requirements:**
- File Form 2553 with IRS
- Deadline: March 15 or within 2.5 months of formation
- Max 100 shareholders
- Only certain shareholders allowed
- One class of stock

### 4. C-Corporation

**Structure:**
- Separate taxpaying entity
- Double taxation
- Most complex structure

**Tax Treatment:**
```
Corporate Level:
Trading income: $500,000
Corporate tax (21%): $105,000
After-tax profit: $395,000

Shareholder Level:
Dividend paid: $200,000
Dividend tax (20%): $40,000
Total tax: $145,000 (29%)

Retained in Corp: $195,000
└─ Can reinvest tax-deferred
```

**Advantages:**
```
✓ Lowest corporate rate (21%)
✓ Retain earnings tax-efficiently
✓ More fringe benefits
✓ Easier to raise capital
✓ Unlimited shareholders
```

**Disadvantages:**
```
✗ Double taxation on dividends
✗ Complex accounting
✗ More regulations
✗ Higher costs
✗ Less flexibility
```

**Best For:**
- Very high income ($500K+)
- Want to retain earnings
- Planning to go public
- Need extensive fringe benefits

### Entity Selection Decision Tree

```
START: Choose Your Trading Entity
│
├─► Trading Income < $50K?
│   └─ YES → Sole Proprietorship
│            (Keep it simple)
│
├─► Trading Income $50K-$150K?
│   └─ YES → LLC (taxed as disregarded)
│            (Liability protection)
│
├─► Trading Income $150K-$300K?
│   ├─ Have other business income?
│   │  └─ YES → LLC taxed as S-Corp
│   │           (SE tax savings on other income)
│   └─ NO → LLC disregarded
│           (Trading = no SE tax anyway)
│
└─► Trading Income > $300K?
    ├─ Want to retain earnings?
    │  └─ YES → C-Corporation
    │           (Lower initial tax)
    └─ NO → LLC or S-Corp
            (Pass-through)
```

## Hands-On Practice: Entity Comparison Calculator (15 minutes)

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def compare_entity_taxes(trading_income, business_expenses, other_income=0):
    """
    Compare tax liability across entity types
    """
    results = {}
    
    net_trading = trading_income - business_expenses
    
    # 1. SOLE PROP / LLC (Disregarded)
    sole_agi = net_trading + other_income
    sole_taxable = max(0, sole_agi - 14600)  # Standard deduction
    sole_qbi = min(sole_taxable * 0.20, net_trading * 0.20)
    sole_tax = (sole_taxable - sole_qbi) * 0.24  # Simplified
    sole_se = other_income * 0.153  # SE only on non-trading income
    
    results['Sole Prop/LLC'] = {
        'Federal Income Tax': sole_tax,
        'SE Tax': sole_se,
        'Total Tax': sole_tax + sole_se,
        'Setup Cost': 100,
        'Annual Cost': 200
    }
    
    # 2. S-CORP (with reasonable salary)
    # Assume 40% of income as W-2 salary for other business income
    if other_income > 0:
        salary = other_income * 0.40
        distribution = net_trading + (other_income - salary)
    else:
        salary = 0
        distribution = net_trading
    
    scorp_agi = salary + distribution
    scorp_taxable = max(0, scorp_agi - 14600)
    scorp_qbi = min(scorp_taxable * 0.20, distribution * 0.20)
    scorp_tax = (scorp_taxable - scorp_qbi) * 0.24
    scorp_se = salary * 0.153  # Only on salary
    scorp_admin = 2000  # Payroll costs
    
    results['S-Corp'] = {
        'Federal Income Tax': scorp_tax,
        'SE Tax': scorp_se,
        'Total Tax': scorp_tax + scorp_se,
        'Setup Cost': 1000,
        'Annual Cost': scorp_admin
    }
    
    # 3. C-CORP (with salary + dividends)
    if other_income > 0:
        c_salary = (net_trading + other_income) * 0.30
    else:
        c_salary = net_trading * 0.30
    
    c_corporate_income = net_trading + other_income - c_salary
    c_corporate_tax = c_corporate_income * 0.21
    
    # Pay out 50% as dividend, retain rest
    c_dividend = (c_corporate_income - c_corporate_tax) * 0.50
    c_dividend_tax = c_dividend * 0.20  # Qualified dividend rate
    
    c_personal_income_tax = c_salary * 0.24
    c_se = c_salary * 0.153
    c_admin = 3000
    
    results['C-Corp'] = {
        'Corporate Tax': c_corporate_tax,
        'Personal Income Tax': c_personal_income_tax,
        'Dividend Tax': c_dividend_tax,
        'SE Tax': c_se,
        'Total Tax': c_corporate_tax + c_personal_income_tax + c_dividend_tax + c_se,
        'Setup Cost': 2000,
        'Annual Cost': c_admin
    }
    
    return results

# Compare scenarios
print("=" * 80)
print("ENTITY COMPARISON: $180K Trading + $0 Other Income")
print("=" * 80)

results1 = compare_entity_taxes(180000, 30000, 0)

for entity, data in results1.items():
    print(f"\n{entity}:")
    for key, value in data.items():
        print(f"  {key:<25} ${value:>10,.0f}")
    
print("\n\n" + "=" * 80)
print("ENTITY COMPARISON: $180K Trading + $60K Consulting")
print("=" * 80)

results2 = compare_entity_taxes(180000, 30000, 60000)

for entity, data in results2.items():
    print(f"\n{entity}:")
    for key, value in data.items():
        print(f"  {key:<25} ${value:>10,.0f}")
````

## Summary

**Entity Selection Guide:**
- **<$50K**: Sole Proprietorship (simplicity)
- **$50K-$150K**: LLC (liability protection)
- **$150K-$300K**: LLC or S-Corp (if other income)
- **>$300K**: Consider C-Corp (if retaining earnings)

**Key Insight**: For pure trading income, entity choice is about liability protection and administrative preference, NOT SE tax (trading exempt).
