### Nexus Rules

**Physical Nexus:**
- Permanent residence
- Physical office
- Employee presence
- Inventory location

**Economic Nexus:**
- Varies by state
- Usually not applicable to traders
- More relevant for businesses

### No-Tax State Benefits

**Texas:**
- No state income tax
- No corporate income tax
- Franchise tax (minimal for small business)
- Major cities: Dallas, Austin, Houston

**Florida:**
- No state income tax
- No corporate income tax
- Business-friendly
- Major cities: Miami, Tampa, Jacksonville

**Nevada:**
- No state income tax
- Privacy protections
- Business-friendly laws
- Lower cost of living

### Relocation Considerations

**Domicile Change Requirements:**
1. Physical presence in new state
2. Change driver's license
3. Register to vote
4. Update address everywhere
5. Establish new banking
6. Spend >183 days in new state

**Exit Tax Considerations:**
- California: May audit last year
- New York: Aggressive residency audits
- Document everything
- Keep day counting records

## Hands-On Practice: State Tax Comparison (15 minutes)

````python
import pandas as pd

# State tax comparison
states = pd.DataFrame({
    'State': ['California', 'New York', 'Texas', 'Florida', 'Nevada'],
    'Income_Tax_Rate': [0.133, 0.109, 0, 0, 0],
    'Property_Tax': [0.73, 1.40, 1.60, 0.83, 0.53],
    'Sales_Tax': [7.25, 4.0, 6.25, 6.0, 6.85]
})

# Calculate taxes on $200K income
income = 200000
states['Income_Tax'] = states['Income_Tax_Rate'] * income

print("Annual Tax Comparison ($200K Income):")
print(states[['State', 'Income_Tax']].to_string(index=False))
print(f"\nMax Savings: ${states['Income_Tax'].max():,.0f}/year")
````
