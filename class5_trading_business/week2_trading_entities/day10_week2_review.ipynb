### Case Study Review

**Case 1: New Trader ($40K)**
- Start: Sole proprietorship
- Reason: Simplicity, low cost
- Future: Form LLC when income >$50K

**Case 2: Full-Time Trader ($180K, CA)**
- Entity: LLC (home state)
- Tax: Disregarded entity
- Consideration: Move to Nevada/Texas
- Potential savings: $24K state tax

**Case 3: Trader + Coach ($200K + $80K)**
- Entity: LLC → S-Corp election
- Salary: $112K (40% of $280K)
- Distribution: $168K
- SE tax savings: ~$18K

**Case 4: High Earner ($500K+)**
- Consider: C-Corporation
- Retain: $200K in corp (21% tax)
- Distribute: $300K (salary + dividend)
- Benefit: Tax deferral, benefits

## Hands-On Practice: Complete Entity Analysis (15 minutes)

````python
import pandas as pd

def complete_entity_recommendation(profile):
    """
    Comprehensive entity recommendation
    """
    trading = profile['trading_income']
    other = profile['other_income']
    state = profile['state']
    total = trading + other
    
    # Entity recommendation
    if total < 50000:
        entity = "Sole Proprietorship"
        reason = "Simplicity and low cost"
    elif total < 150000:
        entity = "LLC (Disregarded)"
        reason = "Liability protection, pass-through tax"
    elif other > 50000:
        entity = "LLC → S-Corp"
        reason = "SE tax savings on non-trading income"
    elif total > 300000:
        entity = "Consider C-Corp"
        reason = "Retain earnings at lower rate"
    else:
        entity = "LLC (Disregarded)"
        reason = "Pure trading, pass-through best"
    
    # State tax analysis
    high_tax_states = ['California', 'New York', 'New Jersey']
    if state in high_tax_states:
        state_rec = f"Consider relocating to save ${total * 0.10:,.0f}/year"
    else:
        state_rec = "State tax optimized"
    
    return {
        'entity': entity,
        'reason': reason,
        'state_rec': state_rec
    }

# Test profiles
profiles = [
    {'name': 'New Trader', 'trading_income': 40000, 'other_income': 0, 'state': 'Texas'},
    {'name': 'Full-Time CA', 'trading_income': 180000, 'other_income': 0, 'state': 'California'},
    {'name': 'Multi-Income', 'trading_income': 200000, 'other_income': 80000, 'state': 'Florida'}
]

for p in profiles:
    result = complete_entity_recommendation(p)
    print(f"\n{p['name']}:")
    print(f"  Entity: {result['entity']}")
    print(f"  Reason: {result['reason']}")
    print(f"  State: {result['state_rec']}")
````
