## Lecture: LLC Formation Process (30 minutes)

### Step-by-Step LLC Formation

```
LLC FORMATION ROADMAP
┌────────────────────────────────────────────────┐
│ 1. Choose State (usually home state)          │
│ 2. Select LLC Name (check availability)       │
│ 3. File Articles of Organization              │
│ 4. Create Operating Agreement                 │
│ 5. Obtain EIN from IRS                        │
│ 6. Open Business Bank Account                 │
│ 7. Setup Accounting System                    │
│ 8. File Annual Reports                        │
└────────────────────────────────────────────────┘
```

### State Selection

**Home State (Recommended for most):**
- Lower fees
- Simpler compliance
- Avoid double filing

**Delaware/Wyoming (Advanced):**
- Business-friendly laws
- Privacy protections
- May require foreign qualification

### Operating Agreement Essentials

**Key Sections:**
1. Member information
2. Capital contributions
3. Profit/loss allocation
4. Management structure
5. Trading authority
6. Distribution rules
7. Dissolution terms

### Compliance Requirements

**Annual:**
- File state annual report
- Pay franchise tax (if applicable)
- Update registered agent
- Maintain separate records

**Ongoing:**
- Keep personal/business separate
- Document major decisions
- Hold annual meetings (recommended)
- Maintain adequate capitalization

## Hands-On Practice: LLC Setup Calculator (15 minutes)

````python
import pandas as pd

# State-by-state LLC costs
state_costs = pd.DataFrame({
    'State': ['California', 'Delaware', 'Florida', 'Nevada', 'Texas', 'Wyoming'],
    'Filing_Fee': [70, 90, 125, 425, 300, 100],
    'Annual_Fee': [800, 300, 139, 350, 0, 60],
    'Franchise_Tax': [800, 0, 0, 0, 0, 0]
})

state_costs['Total_First_Year'] = state_costs['Filing_Fee'] + state_costs['Annual_Fee'] + state_costs['Franchise_Tax']
state_costs['Total_Annual'] = state_costs['Annual_Fee'] + state_costs['Franchise_Tax']

print(state_costs.to_string(index=False))
````
