## Lecture: S-Corporation Benefits (30 minutes)

### S-Corp Tax Structure

```
S-CORP INCOME SPLIT
┌─────────────────────────────────────────┐
│ Total Business Income: $200,000         │
│                                          │
│ REASONABLE SALARY (W-2)                 │
│ ├─ $80,000 (40% of income)             │
│ ├─ Subject to SE tax (15.3%)           │
│ └─ SE Tax: $12,240                     │
│                                          │
│ DISTRIBUTIONS (K-1)                     │
│ ├─ $120,000 (60% of income)            │
│ ├─ NOT subject to SE tax               │
│ └─ SE Tax: $0                          │
│                                          │
│ TOTAL SE TAX: $12,240                  │
│                                          │
│ vs Sole Prop on $200K other income:    │
│ SE Tax would be: $30,600               │
│ SAVINGS: $18,360                       │
└─────────────────────────────────────────┘

NOTE: Pure trading income NOT subject to SE tax,
      so S-Corp mainly beneficial for:
      - Trading education income
      - Consulting/coaching fees
      - Other business activities
```

### Election Process

**Form 2553:**
- File within 2.5 months of formation
- Or by March 15 for existing entity
- All shareholders must consent

### Reasonable Compensation

**IRS Guidelines:**
- Industry standards
- Time/effort invested
- Typically 40-60% of profits
- Cannot be artificially low

### Payroll Requirements

**Must Have:**
- Payroll service or software
- Quarterly 941 filings
- Annual W-2 issuance
- State unemployment tax
- Workers comp (some states)

**Costs:**
- Payroll service: $50-100/month
- Additional accounting: $100-200/month
- Annual: $2,000-3,000

## Hands-On Practice: S-Corp Calculator (15 minutes)

````python
def calculate_scorp_savings(trading_income, other_income, salary_pct=0.40):
    """
    Calculate S-Corp SE tax savings
    Note: Trading income not subject to SE tax
    """
    # Without S-Corp (all other income subject to SE)
    sole_se_tax = min(other_income, 160200) * 0.153  # 2024 limit
    
    # With S-Corp (only salary subject to SE)
    salary = other_income * salary_pct
    distribution = trading_income + (other_income - salary)
    scorp_se_tax = min(salary, 160200) * 0.153
    
    # Costs
    scorp_costs = 2500  # Annual
    
    savings = sole_se_tax - scorp_se_tax - scorp_costs
    
    return {
        'Sole_SE_Tax': sole_se_tax,
        'SCorp_SE_Tax': scorp_se_tax,
        'SCorp_Costs': scorp_costs,
        'Net_Savings': savings,
        'Salary': salary,
        'Distribution': distribution
    }

# Test scenarios
scenarios = [
    {'trading': 150000, 'other': 0, 'name': 'Pure Trading'},
    {'trading': 150000, 'other': 60000, 'name': 'Trading + Coaching'},
    {'trading': 200000, 'other': 100000, 'name': 'Multi-Income'}
]

for s in scenarios:
    result = calculate_scorp_savings(s['trading'], s['other'])
    print(f"\n{s['name']}:")
    print(f"  Net Savings: ${result['Net_Savings']:,.0f}")
    if result['Net_Savings'] > 0:
        print(f"  ✓ S-Corp beneficial")
    else:
        print(f"  ✗ S-Corp not worth it")
````

## Summary

S-Corp election beneficial primarily when you have substantial non-trading business income (coaching, consulting, education). For pure trading, benefit is limited since trading not subject to SE tax.
