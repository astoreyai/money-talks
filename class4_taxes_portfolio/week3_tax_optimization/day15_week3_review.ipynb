### Day 11: Tax-Loss Harvesting Fundamentals

#### Key Concepts

**Tax-Loss Harvesting Definition:**
- Selling securities at a loss to offset capital gains
- Reduces current year tax liability
- Can carry forward unused losses indefinitely

**Annual Deduction Limits:**
```
$3,000 Maximum Annual Deduction Against Ordinary Income
($1,500 if married filing separately)

UNLIMITED offset against capital gains
Unused losses carry forward indefinitely
```

**Loss Priority Hierarchy:**
```
1. Short-term losses offset short-term gains
2. Long-term losses offset long-term gains
3. Net short-term loss offsets long-term gains
4. Net long-term loss offsets short-term gains
5. Remaining loss (up to $3,000) offsets ordinary income
6. Excess losses carry forward
```

**Harvesting Thresholds:**
- Minimum loss threshold: $500-$1,000 (to cover transaction costs)
- Optimal harvesting: Losses > $1,000
- Consider tax bracket and expected gains

**Tax Savings Formula:**
```
Tax Savings = Loss Amount × Tax Rate

Example: $5,000 loss at 24% bracket
= $5,000 × 0.24 = $1,200 tax savings
```

### Day 12: Wash Sale Rules

#### Key Concepts

**Wash Sale Definition:**
- Selling a security at a loss
- Purchasing substantially identical security within 61-day window
- Loss is disallowed for current year

**61-Day Window:**
```
[-----30 days before-----][SALE][-----30 days after-----]
         Window Start              Window End

No repurchase of substantially identical securities allowed
```

**Substantially Identical Securities:**
- Same stock (e.g., AAPL = AAPL)
- Same company bonds
- Options on the same security
- Contracts to acquire the same security

**NOT Substantially Identical:**
- Different companies in same sector (AAPL ≠ MSFT)
- Different index funds tracking different indexes
- Similar but not identical ETFs
- Preferred vs common stock

**Legal Tax-Loss Harvesting Swaps:**

| Sell | Buy | Valid? |
|------|-----|--------|
| VOO (S&P 500) | IVV (S&P 500) | YES - different issuers |
| VTI (Total Market) | SCHB (Total Market) | YES - different issuers |
| AAPL | MSFT | YES - different companies |
| BND (Bond Fund) | AGG (Bond Fund) | YES - different holdings |
| AAPL | AAPL | NO - same security |
| VTI | VTSAX | MAYBE - same underlying |

**Wash Sale Cost Basis Adjustment:**
```
When wash sale occurs:
1. Disallowed loss added to cost basis of new purchase
2. Holding period includes original purchase date
3. Loss deferred, not permanently lost

Example:
Sold 100 shares at $2,000 loss (wash sale)
Repurchased 100 shares at $8,000
New cost basis: $8,000 + $2,000 = $10,000
```

**Avoiding Wash Sales:**
1. Wait 31 days before repurchasing
2. Buy similar but not identical security
3. Use double-up strategy (buy first, sell 31+ days later)
4. Harvest losses in taxable accounts only
5. Track purchases across all accounts (IRA purchases trigger wash sales)

**Tax-Deferred (401k/Traditional IRA) - Best for:**

**Asset Location Decision Framework:**
```
FOR EACH INVESTMENT:
```

### Day 14: Tax-Efficient Funds

#### Key Concepts

**Tax Efficiency Metrics:**

**1. Turnover Ratio:**
```
Turnover Ratio = Value of Holdings Sold / Average Portfolio Value

Classification:
- Low turnover: <20% (1-2 distributions/year)
- Moderate: 20-50%
- High turnover: >50% (frequent distributions)
```

**2. Tax-Cost Ratio:**
```
Tax-Cost Ratio = Annual Tax Impact / Pre-Tax Return

Example:
Fund returns 10% before taxes
Taxes reduce return to 8.5%
Tax-Cost Ratio = 1.5%
```

**3. Qualified Dividend Income (QDI):**
```
Percentage of dividends taxed at preferential rates

High QDI (>80%): Tax-efficient
Low QDI (<50%): Less tax-efficient
```

**ETFs vs Mutual Funds - Tax Comparison:**

| Feature | ETFs | Mutual Funds |
|---------|------|-------------|
| Capital gains distributions | Rare | Common |
| Creation/redemption | In-kind (tax-efficient) | Cash (triggers gains) |
| Investor control | Timing flexibility | Fund manager driven |
| Tax efficiency | High | Varies |
| Average turnover | Lower | Higher |

**Why ETFs Are More Tax-Efficient:**
```
1. In-Kind Creation/Redemption
   - No taxable event when shares created/redeemed
   - Authorized Participants exchange securities directly

2. Authorized Participant (AP) Mechanism
   - APs redeem shares using low-cost-basis securities
   - Increases ETF's average cost basis
   - Minimizes embedded capital gains

3. Investor Control
   - You choose when to sell (realize gains)
   - Not affected by other investors' redemptions
```

**Most Tax-Efficient Fund Types:**

1. **Broad Market Index ETFs**
   - Examples: VOO, VTI, IVV
   - Turnover: <5%
   - QDI: >80%

2. **Municipal Bond Funds**
   - Interest is tax-exempt
   - Low turnover
   - For taxable accounts

3. **Tax-Managed Funds**
   - Explicitly designed for tax efficiency
   - Active tax-loss harvesting
   - Minimize distributions

**Least Tax-Efficient Fund Types:**

1. **Actively Managed Stock Funds**
   - Turnover: 50-200%
   - Frequent short-term gains

2. **Bond Funds (Taxable)**
   - Interest taxed as ordinary income
   - No preferential rates

3. **Sector Funds**
   - High turnover
   - Concentrated positions

## Part 3: Python Code - Comprehensive Tax Optimization Tools

````python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
from typing import Dict, List, Tuple
import warnings
warnings.filterwarnings('ignore')

print("Tax Optimization Tools Loaded Successfully")
print("="*60)
````

### Tool 1: Comprehensive Tax Optimization Analyzer

````python
class TaxOptimizationAnalyzer:
    """
    Comprehensive tax optimization analyzer integrating all Week 3 concepts:
    - Tax-loss harvesting identification
    - Wash sale checking
    - Asset location optimization
    - Fund tax efficiency analysis
    """
    
    def __init__(self, tax_bracket: float):
        """
        Initialize analyzer with investor's tax bracket
        
        Parameters:
        tax_bracket: Marginal tax rate (e.g., 0.24 for 24%)
        """
        self.tax_bracket = tax_bracket
        self.ltcg_rate = 0.15 if tax_bracket <= 0.24 else 0.20
        self.wash_sale_window = 61  # days
        self.min_harvest_threshold = 1000  # minimum loss to harvest
        
    def analyze_harvest_opportunities(self, portfolio: pd.DataFrame) -> pd.DataFrame:
        """
        Identify tax-loss harvesting opportunities
        
        Parameters:
        portfolio: DataFrame with columns ['ticker', 'shares', 'cost_basis', 'current_price', 'purchase_date']
        
        Returns:
        DataFrame of harvest opportunities sorted by tax benefit
        """
        portfolio = portfolio.copy()
        
        # Calculate unrealized gains/losses
        portfolio['total_cost'] = portfolio['shares'] * portfolio['cost_basis']
        portfolio['current_value'] = portfolio['shares'] * portfolio['current_price']
        portfolio['unrealized_gl'] = portfolio['current_value'] - portfolio['total_cost']
        portfolio['gl_percent'] = (portfolio['unrealized_gl'] / portfolio['total_cost']) * 100
        
        # Determine holding period
        portfolio['purchase_date'] = pd.to_datetime(portfolio['purchase_date'])
        portfolio['days_held'] = (datetime.now() - portfolio['purchase_date']).dt.days
        portfolio['holding_period'] = portfolio['days_held'].apply(
            lambda x: 'Long-term' if x > 365 else 'Short-term'
        )
        
        # Calculate tax rate and potential savings
        portfolio['tax_rate'] = portfolio['holding_period'].apply(
            lambda x: self.ltcg_rate if x == 'Long-term' else self.tax_bracket
        )
        portfolio['tax_savings'] = -portfolio['unrealized_gl'] * portfolio['tax_rate']
        
        # Filter for harvest opportunities
        harvest_opps = portfolio[
            (portfolio['unrealized_gl'] < -self.min_harvest_threshold)
        ].copy()
        
        # Calculate priority score (higher is better)
        # Prioritize: 1) Short-term losses, 2) Larger tax savings, 3) Larger loss percentage
        harvest_opps['priority_score'] = (
            (harvest_opps['holding_period'] == 'Short-term').astype(int) * 1000 +
            harvest_opps['tax_savings'] +
            abs(harvest_opps['gl_percent'])
        )
        
        # Sort by priority
        harvest_opps = harvest_opps.sort_values('priority_score', ascending=False)
        
        return harvest_opps[[
            'ticker', 'shares', 'cost_basis', 'current_price', 'unrealized_gl',
            'gl_percent', 'holding_period', 'tax_rate', 'tax_savings', 'priority_score'
        ]]
    
    def check_wash_sales(self, 
                        sell_ticker: str, 
                        sell_date: datetime,
                        transaction_history: pd.DataFrame) -> Dict:
        """
        Check for potential wash sale violations
        
        Parameters:
        sell_ticker: Ticker symbol being sold
        sell_date: Date of sale
        transaction_history: DataFrame with columns ['ticker', 'date', 'action', 'shares']
        
        Returns:
        Dictionary with wash sale analysis
        """
        window_start = sell_date - timedelta(days=30)
        window_end = sell_date + timedelta(days=30)
        
        # Find purchases of same security in window
        transaction_history['date'] = pd.to_datetime(transaction_history['date'])
        
        purchases_in_window = transaction_history[
            (transaction_history['ticker'] == sell_ticker) &
            (transaction_history['action'] == 'BUY') &
            (transaction_history['date'] >= window_start) &
            (transaction_history['date'] <= window_end) &
            (transaction_history['date'] != sell_date)
        ]
        
        is_wash_sale = len(purchases_in_window) > 0
        
        return {
            'is_wash_sale': is_wash_sale,
            'window_start': window_start,
            'window_end': window_end,
            'violating_purchases': purchases_in_window,
            'safe_repurchase_date': window_end + timedelta(days=1),
            'recommendation': 'WAIT' if is_wash_sale else 'CLEAR'
        }
    
    def suggest_swap_alternatives(self, ticker: str) -> List[str]:
        """
        Suggest alternative securities for tax-loss harvesting swaps
        
        Parameters:
        ticker: Original ticker symbol
        
        Returns:
        List of alternative ticker symbols
        """
        # Common swap alternatives (simplified for education)
        swap_map = {
            # S&P 500 ETFs
            'VOO': ['IVV', 'SPY'],
            'IVV': ['VOO', 'SPY'],
            'SPY': ['VOO', 'IVV'],
            
            # Total Market ETFs
            'VTI': ['ITOT', 'SCHB'],
            'ITOT': ['VTI', 'SCHB'],
            'SCHB': ['VTI', 'ITOT'],
            
            # Bond ETFs
            'BND': ['AGG', 'SCHZ'],
            'AGG': ['BND', 'SCHZ'],
            'SCHZ': ['BND', 'AGG'],
            
            # International
            'VXUS': ['IXUS', 'SCHF'],
            'IXUS': ['VXUS', 'SCHF'],
            'SCHF': ['VXUS', 'IXUS'],
        }
        
        return swap_map.get(ticker, ['No common alternatives - consider similar sector'])
    
    def analyze_asset_location(self, holdings: pd.DataFrame) -> pd.DataFrame:
        """
        Analyze asset location efficiency and suggest improvements
        
        Parameters:
        holdings: DataFrame with columns ['ticker', 'asset_type', 'account_type', 'value']
        
        Returns:
        DataFrame with location analysis and recommendations
        """
        holdings = holdings.copy()
        
        # Define optimal account types for each asset type
        optimal_location = {
            'Municipal Bonds': 'Taxable',
            'Index Fund': 'Taxable',
            'Large-Cap Stock': 'Taxable',
            'Foreign Stock': 'Taxable',
            'Small-Cap Stock': 'Roth',
            'Emerging Markets': 'Roth',
            'Taxable Bonds': 'Tax-Deferred',
            'REIT': 'Tax-Deferred',
            'Active Fund': 'Tax-Deferred',
            'High-Yield Bonds': 'Tax-Deferred',
        }
        
        # Tax efficiency scores (higher = more tax-efficient)
        tax_efficiency = {
            'Municipal Bonds': 10,
            'Index Fund': 9,
            'Large-Cap Stock': 8,
            'Foreign Stock': 7,
            'Small-Cap Stock': 6,
            'Active Fund': 4,
            'Taxable Bonds': 3,
            'REIT': 2,
            'High-Yield Bonds': 1,
        }
        
        holdings['optimal_account'] = holdings['asset_type'].map(optimal_location)
        holdings['tax_efficiency'] = holdings['asset_type'].map(tax_efficiency)
        holdings['correctly_located'] = holdings['account_type'] == holdings['optimal_account']
        
        # Calculate potential tax savings from relocation
        holdings['annual_tax_drag'] = holdings.apply(
            lambda row: 0 if row['correctly_located'] else 
            row['value'] * 0.02 * (10 - row['tax_efficiency']) / 10,
            axis=1
        )
        
        holdings['recommendation'] = holdings.apply(
            lambda row: 'OPTIMAL' if row['correctly_located'] else 
            f"Consider moving to {row['optimal_account']}",
            axis=1
        )
        
        return holdings[[
            'ticker', 'asset_type', 'account_type', 'value',
            'optimal_account', 'tax_efficiency', 'correctly_located',
            'annual_tax_drag', 'recommendation'
        ]].sort_values('annual_tax_drag', ascending=False)
    
    def analyze_fund_efficiency(self, funds: pd.DataFrame) -> pd.DataFrame:
        """
        Analyze tax efficiency of funds
        
        Parameters:
        funds: DataFrame with columns ['ticker', 'type', 'turnover', 'qdi_percent', 'expense_ratio']
        
        Returns:
        DataFrame with tax efficiency analysis
        """
        funds = funds.copy()
        
        # Calculate tax efficiency score (0-100)
        # Lower turnover = better
        # Higher QDI = better
        # ETF structure = better
        
        funds['turnover_score'] = funds['turnover'].apply(
            lambda x: 100 if x < 20 else (80 if x < 50 else 50 if x < 100 else 20)
        )
        
        funds['qdi_score'] = funds['qdi_percent']
        
        funds['structure_score'] = funds['type'].apply(
            lambda x: 100 if x == 'ETF' else 70
        )
        
        funds['tax_efficiency_score'] = (
            funds['turnover_score'] * 0.4 +
            funds['qdi_score'] * 0.3 +
            funds['structure_score'] * 0.3
        )
        
        # Estimate tax-cost ratio
        funds['estimated_tax_cost'] = funds.apply(
            lambda row: (
                row['turnover'] / 100 * self.ltcg_rate * 0.5 +  # Capital gains
                (100 - row['qdi_percent']) / 100 * self.tax_bracket * 0.02  # Non-qualified dividends
            ),
            axis=1
        )
        
        # Classification
        funds['efficiency_rating'] = funds['tax_efficiency_score'].apply(
            lambda x: 'Excellent' if x >= 85 else
                     'Good' if x >= 70 else
                     'Fair' if x >= 50 else
                     'Poor'
        )
        
        funds['recommendation'] = funds.apply(
            lambda row: 'Keep in taxable account' if row['tax_efficiency_score'] >= 70 else
                       'Consider for tax-deferred account' if row['tax_efficiency_score'] >= 50 else
                       'Move to tax-deferred or replace with ETF',
            axis=1
        )
        
        return funds[[
            'ticker', 'type', 'turnover', 'qdi_percent',
            'tax_efficiency_score', 'estimated_tax_cost',
            'efficiency_rating', 'recommendation'
        ]].sort_values('tax_efficiency_score', ascending=False)
    
    def calculate_total_tax_impact(self, 
                                   realized_stcg: float,
                                   realized_ltcg: float,
                                   harvested_losses: float,
                                   qualified_dividends: float,
                                   ordinary_dividends: float) -> Dict:
        """
        Calculate comprehensive tax impact for the year
        
        Returns:
        Dictionary with detailed tax calculations
        """
        # Apply losses using IRS ordering rules
        remaining_losses = abs(harvested_losses)
        
        # 1. Offset short-term gains
        stcg_after_loss = max(0, realized_stcg - remaining_losses)
        remaining_losses = max(0, remaining_losses - realized_stcg)
        
        # 2. Offset long-term gains
        ltcg_after_loss = max(0, realized_ltcg - remaining_losses)
        remaining_losses = max(0, remaining_losses - realized_ltcg)
        
        # 3. Offset ordinary income (max $3,000)
        ordinary_income_offset = min(3000, remaining_losses)
        remaining_losses = max(0, remaining_losses - 3000)
        
        # Calculate taxes
        stcg_tax = stcg_after_loss * self.tax_bracket
        ltcg_tax = ltcg_after_loss * self.ltcg_rate
        qualified_div_tax = qualified_dividends * self.ltcg_rate
        ordinary_div_tax = ordinary_dividends * self.tax_bracket
        ordinary_income_tax_savings = ordinary_income_offset * self.tax_bracket
        
        total_investment_tax = (
            stcg_tax + ltcg_tax + qualified_div_tax + ordinary_div_tax -
            ordinary_income_tax_savings
        )
        
        # Calculate tax savings from harvesting
        tax_without_harvesting = (
            realized_stcg * self.tax_bracket +
            realized_ltcg * self.ltcg_rate +
            qualified_dividends * self.ltcg_rate +
            ordinary_dividends * self.tax_bracket
        )
        
        tax_savings = tax_without_harvesting - total_investment_tax
        
        return {
            'realized_stcg': realized_stcg,
            'realized_ltcg': realized_ltcg,
            'harvested_losses': harvested_losses,
            'stcg_after_loss': stcg_after_loss,
            'ltcg_after_loss': ltcg_after_loss,
            'ordinary_income_offset': ordinary_income_offset,
            'loss_carryforward': remaining_losses,
            'stcg_tax': stcg_tax,
            'ltcg_tax': ltcg_tax,
            'qualified_div_tax': qualified_div_tax,
            'ordinary_div_tax': ordinary_div_tax,
            'total_investment_tax': total_investment_tax,
            'tax_savings_from_harvesting': tax_savings,
            'effective_tax_rate': total_investment_tax / (
                realized_stcg + realized_ltcg + qualified_dividends + ordinary_dividends
            ) if (realized_stcg + realized_ltcg + qualified_dividends + ordinary_dividends) > 0 else 0
        }

print("✓ TaxOptimizationAnalyzer class defined")
print("  Methods: analyze_harvest_opportunities, check_wash_sales,")
print("           suggest_swap_alternatives, analyze_asset_location,")
print("           analyze_fund_efficiency, calculate_total_tax_impact")
````

### Tool 2: Year-End Tax Planning Tool

````python
class YearEndTaxPlanner:
    """
    Year-end tax planning tool for strategic decision-making
    """
    
    def __init__(self, current_year_income: float, tax_bracket: float):
        self.current_year_income = current_year_income
        self.tax_bracket = tax_bracket
        self.ltcg_rate = 0.15 if tax_bracket <= 0.24 else 0.20
        
    def project_year_end_taxes(self,
                               ytd_stcg: float,
                               ytd_ltcg: float,
                               ytd_losses: float,
                               potential_harvests: float,
                               qualified_dividends: float,
                               ordinary_dividends: float) -> pd.DataFrame:
        """
        Project year-end tax scenarios
        
        Returns:
        DataFrame comparing scenarios with and without additional harvesting
        """
        scenarios = []
        
        # Scenario 1: Do nothing
        scenario1 = self._calculate_scenario(
            'Do Nothing',
            ytd_stcg, ytd_ltcg, ytd_losses,
            qualified_dividends, ordinary_dividends
        )
        scenarios.append(scenario1)
        
        # Scenario 2: Harvest available losses
        scenario2 = self._calculate_scenario(
            'Harvest Available Losses',
            ytd_stcg, ytd_ltcg, ytd_losses + potential_harvests,
            qualified_dividends, ordinary_dividends
        )
        scenarios.append(scenario2)
        
        # Scenario 3: Harvest losses + realize some gains (if in low income year)
        if self.tax_bracket <= 0.12:
            # Can realize up to LTCG limit at 0% rate
            additional_gains = 50000  # Example
            scenario3 = self._calculate_scenario(
                'Harvest + Realize Gains (0% rate)',
                ytd_stcg, ytd_ltcg + additional_gains, ytd_losses + potential_harvests,
                qualified_dividends, ordinary_dividends
            )
            scenarios.append(scenario3)
        
        df = pd.DataFrame(scenarios)
        df['tax_savings'] = df['total_tax'].iloc[0] - df['total_tax']
        
        return df
    
    def _calculate_scenario(self, name: str,
                           stcg: float, ltcg: float, losses: float,
                           qualified_div: float, ordinary_div: float) -> Dict:
        """
        Calculate taxes for a specific scenario
        """
        # Apply losses
        remaining_losses = abs(losses)
        
        stcg_after = max(0, stcg - remaining_losses)
        remaining_losses = max(0, remaining_losses - stcg)
        
        ltcg_after = max(0, ltcg - remaining_losses)
        remaining_losses = max(0, remaining_losses - ltcg)
        
        ordinary_offset = min(3000, remaining_losses)
        carryforward = max(0, remaining_losses - 3000)
        
        # Calculate taxes
        stcg_tax = stcg_after * self.tax_bracket
        ltcg_tax = ltcg_after * self.ltcg_rate
        qual_div_tax = qualified_div * self.ltcg_rate
        ord_div_tax = ordinary_div * self.tax_bracket
        ordinary_savings = ordinary_offset * self.tax_bracket
        
        total_tax = stcg_tax + ltcg_tax + qual_div_tax + ord_div_tax - ordinary_savings
        
        return {
            'scenario': name,
            'stcg': stcg,
            'ltcg': ltcg,
            'losses': losses,
            'stcg_after_loss': stcg_after,
            'ltcg_after_loss': ltcg_after,
            'ordinary_offset': ordinary_offset,
            'carryforward': carryforward,
            'total_tax': total_tax
        }
    
    def create_action_plan(self, days_until_year_end: int,
                          harvest_opportunities: pd.DataFrame) -> List[Dict]:
        """
        Create prioritized action plan for year-end
        
        Returns:
        List of actions with timing and priority
        """
        actions = []
        
        # High priority: Time-sensitive actions
        if days_until_year_end <= 30:
            actions.append({
                'priority': 'CRITICAL',
                'action': 'Execute tax-loss harvesting',
                'deadline': 'Dec 31',
                'details': f"{len(harvest_opportunities)} opportunities identified",
                'tax_savings': f"${harvest_opportunities['tax_savings'].sum():,.2f}"
            })
            
            actions.append({
                'priority': 'CRITICAL',
                'action': 'Avoid mutual fund purchases',
                'deadline': 'Dec 31',
                'details': 'Wait until January to avoid year-end distributions',
                'tax_savings': 'Variable'
            })
        
        # Medium priority: Strategic actions
        actions.append({
            'priority': 'HIGH',
            'action': 'Review asset location',
            'deadline': 'Before year-end',
            'details': 'Optimize placement for next year contributions',
            'tax_savings': 'Long-term benefit'
        })
        
        actions.append({
            'priority': 'HIGH',
            'action': 'Max out retirement contributions',
            'deadline': 'Dec 31 (401k) / Apr 15 (IRA)',
            'details': 'Reduce taxable income',
            'tax_savings': f"${self.tax_bracket * 22500:,.2f} (401k max)"
        })
        
        # Lower priority: Documentation
        actions.append({
            'priority': 'MEDIUM',
            'action': 'Document all transactions',
            'deadline': 'Before filing',
            'details': 'Maintain records for wash sales and cost basis',
            'tax_savings': 'Avoid penalties'
        })
        
        actions.append({
            'priority': 'MEDIUM',
            'action': 'Review charitable giving',
            'deadline': 'Dec 31',
            'details': 'Consider donating appreciated stock',
            'tax_savings': 'Deduction + avoid capital gains'
        })
        
        return actions

print("✓ YearEndTaxPlanner class defined")
print("  Methods: project_year_end_taxes, create_action_plan")
````

### Tool 3: Tax Savings Estimator

````python
def estimate_tax_savings(portfolio_value: float,
                        tax_bracket: float,
                        current_efficiency: str,
                        years: int = 30) -> pd.DataFrame:
    """
    Estimate long-term tax savings from optimization strategies
    
    Parameters:
    portfolio_value: Current portfolio value
    tax_bracket: Marginal tax rate
    current_efficiency: 'poor', 'fair', 'good', 'excellent'
    years: Investment time horizon
    
    Returns:
    DataFrame showing projected savings over time
    """
    # Tax drag estimates by efficiency level
    tax_drag = {
        'poor': 0.025,      # 2.5% annual drag
        'fair': 0.015,      # 1.5% annual drag
        'good': 0.008,      # 0.8% annual drag
        'excellent': 0.003  # 0.3% annual drag
    }
    
    current_drag = tax_drag[current_efficiency]
    optimized_drag = tax_drag['excellent']
    
    annual_return = 0.08  # 8% assumed return
    
    results = []
    
    current_value = portfolio_value
    optimized_value = portfolio_value
    
    for year in range(1, years + 1):
        # Current strategy
        current_value *= (1 + annual_return - current_drag)
        
        # Optimized strategy
        optimized_value *= (1 + annual_return - optimized_drag)
        
        # Calculate difference
        difference = optimized_value - current_value
        difference_pct = (difference / current_value) * 100
        
        results.append({
            'year': year,
            'current_strategy': current_value,
            'optimized_strategy': optimized_value,
            'additional_wealth': difference,
            'improvement_pct': difference_pct
        })
    
    return pd.DataFrame(results)

def summarize_optimization_benefits(tax_bracket: float) -> pd.DataFrame:
    """
    Summarize benefits of each optimization strategy
    
    Returns:
    DataFrame with strategy benefits
    """
    ltcg_rate = 0.15 if tax_bracket <= 0.24 else 0.20
    
    strategies = [
        {
            'strategy': 'Tax-Loss Harvesting',
            'annual_benefit': '0.5% - 1.5% of portfolio',
            'complexity': 'Medium',
            'time_required': '2-4 hours/year',
            'best_for': 'Taxable accounts with $50k+',
            'key_consideration': 'Avoid wash sales'
        },
        {
            'strategy': 'Asset Location',
            'annual_benefit': '0.3% - 1.0% of portfolio',
            'complexity': 'Low',
            'time_required': '1-2 hours/year',
            'best_for': 'Multiple account types',
            'key_consideration': 'Maintain allocation'
        },
        {
            'strategy': 'Tax-Efficient Funds (ETFs)',
            'annual_benefit': '0.2% - 0.8% of portfolio',
            'complexity': 'Low',
            'time_required': '1 hour initial',
            'best_for': 'All taxable accounts',
            'key_consideration': 'Transaction costs to switch'
        },
        {
            'strategy': 'Qualified Dividends',
            'annual_benefit': f'{(tax_bracket - ltcg_rate)*100:.1f}% on dividends',
            'complexity': 'Low',
            'time_required': 'Automatic',
            'best_for': 'Dividend-paying stocks',
            'key_consideration': 'Holding period requirements'
        },
        {
            'strategy': 'Loss Carryforward',
            'annual_benefit': '$3,000 ordinary offset',
            'complexity': 'Low',
            'time_required': 'Automatic',
            'best_for': 'Large prior losses',
            'key_consideration': 'Can last indefinitely'
        }
    ]
    
    return pd.DataFrame(strategies)

print("✓ Tax savings estimation functions defined")
print("  Functions: estimate_tax_savings, summarize_optimization_benefits")
````
