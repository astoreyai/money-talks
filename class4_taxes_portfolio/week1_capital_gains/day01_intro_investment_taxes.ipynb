## Lecture: Introduction to Investment Taxes (30 min)

### Why Taxes Matter

```
THE TAX IMPACT ON RETURNS
=========================

Pre-Tax Return: 10%
Tax Rate: 30%
After-Tax Return: 10% × (1 - 0.30) = 7%

Over 30 years on $100,000:

Pre-Tax Growth:   $100,000 → $1,744,940  (10% annual)
After-Tax Growth: $100,000 → $761,225    (7% annual)
                            ─────────────
Tax Cost:                   $983,715 LOST TO TAXES!

This is why tax-efficient investing matters!
```

### Realized vs Unrealized Gains

```
TAXABLE EVENT TIMING
====================

UNREALIZED GAIN (Paper Profit)
──────────────────────────────
• You bought stock at $50
• Stock is now worth $100
• You have $50 unrealized gain
• NO TAX DUE (you haven't sold)

          ┌─────────────┐
          │   HOLDING   │
          │   No Tax    │
          └─────────────┘
                │
                ▼
         ┌──────────────┐
         │    SELL      │ ← TAXABLE EVENT
         └──────────────┘
                │
                ▼
          ┌─────────────┐
          │ REALIZED    │
          │ GAIN/LOSS   │
          │ Tax Due     │
          └─────────────┘

REALIZED GAIN
─────────────
• You sell the stock for $100
• Your cost basis was $50
• Realized gain: $100 - $50 = $50
• TAX IS NOW DUE on the $50 gain

KEY INSIGHT: You control WHEN to pay taxes by deciding WHEN to sell!
```

### Capital Gains Overview

```
CAPITAL GAINS TAX BASICS
========================

Capital Gain = Sale Price - Cost Basis

HOLDING PERIOD MATTERS:
───────────────────────

├── Taxed as ORDINARY INCOME
├── Rates: 10% to 37% (2024)

LONG-TERM (held > 1 year)
├── PREFERENTIAL tax rates
```

### The Power of Tax Deferral

```
TAX DEFERRAL COMPOUND EFFECT
============================

Scenario: $10,000 investment, 8% annual return, 25% tax rate

STRATEGY A: Sell and pay taxes annually
──────────────────────────────────────
Year 1: $10,000 × 8% = $800 gain
        Tax: $800 × 25% = $200
        Reinvest: $10,000 + $800 - $200 = $10,600
        
After-tax return: 6% per year
After 20 years: $32,071

STRATEGY B: Hold and defer taxes
────────────────────────────────
Year 1-20: Compound at full 8%
After 20 years (pre-tax): $46,610
Tax at sale: ($46,610 - $10,000) × 15% LTCG = $5,491
After-tax: $46,610 - $5,491 = $41,119

BENEFIT OF DEFERRAL: $41,119 - $32,071 = $9,048 MORE!
                    (28% more wealth)

WHY IT WORKS:
• Unrealized gains compound tax-free
• Long-term rates lower than ordinary
• One tax event instead of many
```

## Hands-On Practice: Tax Impact Calculator (15 min)

Let's build tools to calculate the impact of taxes on investment returns.

````python
# Install and import required libraries
!pip install pandas numpy matplotlib -q

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

print("Day 1: Introduction to Investment Taxes")
print("="*45)
````

````python
# 2024 Tax Rate Tables
TAX_BRACKETS_2024 = {
    'single': [
        (11600, 0.10),
        (47150, 0.12),
        (100525, 0.22),
        (191950, 0.24),
        (243725, 0.32),
        (609350, 0.35),
        (float('inf'), 0.37)
    ],
    'married': [
        (23200, 0.10),
        (94300, 0.12),
        (201050, 0.22),
        (383900, 0.24),
        (487450, 0.32),
        (731200, 0.35),
        (float('inf'), 0.37)
    ]
}

LTCG_BRACKETS_2024 = {
    'single': [
        (47025, 0.00),
        (518900, 0.15),
        (float('inf'), 0.20)
    ],
    'married': [
        (94050, 0.00),
        (583750, 0.15),
        (float('inf'), 0.20)
    ]
}

def calculate_ordinary_tax(income, filing_status='single'):
    """Calculate tax on ordinary income."""
    brackets = TAX_BRACKETS_2024[filing_status]
    tax = 0
    prev_limit = 0
    
    for limit, rate in brackets:
        if income <= limit:
            tax += (income - prev_limit) * rate
            break
        else:
            tax += (limit - prev_limit) * rate
            prev_limit = limit
    
    return tax

def get_ltcg_rate(income, filing_status='single'):
    """Get long-term capital gains rate based on income."""
    brackets = LTCG_BRACKETS_2024[filing_status]
    
    for limit, rate in brackets:
        if income <= limit:
            return rate
    
    return brackets[-1][1]

print("Tax rate functions defined!")
print("\nExample: Single filer with $75,000 income")
print(f"Ordinary income tax: ${calculate_ordinary_tax(75000):,.2f}")
print(f"LTCG rate: {get_ltcg_rate(75000)*100:.0f}%")
````

````python
# Capital Gains Calculator
def calculate_capital_gain_tax(sale_price, cost_basis, holding_days, 
                                ordinary_income, filing_status='single'):
    """
    Calculate tax on a capital gain.
    
    Parameters:
    - sale_price: Amount received from sale
    - cost_basis: Original purchase price + fees
    - holding_days: Days held
    - ordinary_income: Other taxable income
    - filing_status: 'single' or 'married'
    """
    gain = sale_price - cost_basis
    
    if gain <= 0:
        return {
            'gain': gain,
            'tax': 0,
            'rate': 0,
            'type': 'Capital Loss',
            'after_tax_proceeds': sale_price
        }
    
    # Determine if short-term or long-term
    is_long_term = holding_days > 365
    
    if is_long_term:
        # Long-term capital gains rate
        total_income = ordinary_income + gain
        rate = get_ltcg_rate(total_income, filing_status)
        tax_type = 'Long-Term'
    else:
        # Short-term = ordinary income rate
        # Calculate marginal rate
        tax_without = calculate_ordinary_tax(ordinary_income, filing_status)
        tax_with = calculate_ordinary_tax(ordinary_income + gain, filing_status)
        rate = (tax_with - tax_without) / gain if gain > 0 else 0
        tax_type = 'Short-Term'
    
    tax = gain * rate
    
    return {
        'gain': gain,
        'tax': tax,
        'rate': rate,
        'type': tax_type,
        'after_tax_proceeds': sale_price - tax,
        'effective_gain': gain - tax
    }

# Example
print("CAPITAL GAINS TAX EXAMPLE")
print("="*50)
print("\nScenario: Bought stock for $10,000, sold for $15,000")
print("Other income: $80,000")

# Short-term (6 months)
st_result = calculate_capital_gain_tax(15000, 10000, 180, 80000)
print(f"\nShort-Term (held 6 months):")
print(f"  Gain: ${st_result['gain']:,.2f}")
print(f"  Tax Rate: {st_result['rate']*100:.1f}%")
print(f"  Tax: ${st_result['tax']:,.2f}")
print(f"  After-Tax Profit: ${st_result['effective_gain']:,.2f}")

# Long-term (2 years)
lt_result = calculate_capital_gain_tax(15000, 10000, 730, 80000)
print(f"\nLong-Term (held 2 years):")
print(f"  Gain: ${lt_result['gain']:,.2f}")
print(f"  Tax Rate: {lt_result['rate']*100:.1f}%")
print(f"  Tax: ${lt_result['tax']:,.2f}")
print(f"  After-Tax Profit: ${lt_result['effective_gain']:,.2f}")

print(f"\nTax Savings from Holding: ${st_result['tax'] - lt_result['tax']:,.2f}")
````

````python
# Tax Deferral Comparison
def compare_tax_strategies(initial, annual_return, years, 
                           annual_tax_rate, final_tax_rate):
    """
    Compare annual realization vs long-term deferral strategies.
    """
    # Strategy 1: Realize gains annually
    after_tax_return = annual_return * (1 - annual_tax_rate)
    annual_value = initial * (1 + after_tax_return) ** years
    
    # Strategy 2: Defer until end
    pre_tax_value = initial * (1 + annual_return) ** years
    gain = pre_tax_value - initial
    tax_at_end = gain * final_tax_rate
    deferred_value = pre_tax_value - tax_at_end
    
    return {
        'annual_realization': annual_value,
        'deferred': deferred_value,
        'benefit': deferred_value - annual_value,
        'benefit_pct': (deferred_value / annual_value - 1) * 100
    }

# Compare strategies
print("TAX DEFERRAL COMPARISON")
print("="*50)
print("\nAssumptions:")
print("  Initial investment: $100,000")
print("  Annual return: 8%")
print("  Annual tax rate (ST): 24%")
print("  Final tax rate (LT): 15%")

results_by_year = []
for years in [5, 10, 15, 20, 25, 30]:
    result = compare_tax_strategies(100000, 0.08, years, 0.24, 0.15)
    results_by_year.append({
        'Years': years,
        'Annual Realization': result['annual_realization'],
        'Tax Deferred': result['deferred'],
        'Benefit': result['benefit'],
        'Benefit %': result['benefit_pct']
    })

df = pd.DataFrame(results_by_year)
print("\n")
print(df.to_string(index=False, float_format=lambda x: f"${x:,.0f}" if x > 100 else f"{x:.1f}%"))
````

````python
# Visualize Tax Deferral Impact
years = list(range(1, 31))
annual_values = []
deferred_values = []

for y in years:
    result = compare_tax_strategies(100000, 0.08, y, 0.24, 0.15)
    annual_values.append(result['annual_realization'])
    deferred_values.append(result['deferred'])

plt.figure(figsize=(12, 5))

# Plot 1: Growth comparison
plt.subplot(1, 2, 1)
plt.plot(years, annual_values, 'r-', label='Annual Realization', linewidth=2)
plt.plot(years, deferred_values, 'g-', label='Tax Deferred', linewidth=2)
plt.fill_between(years, annual_values, deferred_values, alpha=0.3, color='green')
plt.xlabel('Years')
plt.ylabel('Portfolio Value ($)')
plt.title('Tax Deferral Impact on Portfolio Growth')
plt.legend()
plt.grid(True, alpha=0.3)
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x/1000:.0f}K'))

# Plot 2: Benefit of deferral
plt.subplot(1, 2, 2)
benefits = [d - a for d, a in zip(deferred_values, annual_values)]
plt.bar(years, benefits, color='green', alpha=0.7)
plt.xlabel('Years')
plt.ylabel('Benefit ($)')
plt.title('Dollar Benefit of Tax Deferral')
plt.grid(True, alpha=0.3, axis='y')
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x/1000:.0f}K'))

plt.tight_layout()
plt.show()

print(f"\nAfter 30 years, tax deferral provides ${benefits[-1]:,.0f} more wealth!")
````

````python
# Investment Income Summary Tool
def analyze_investment_income(trades, dividends, interest, 
                               ordinary_income, filing_status='single'):
    """
    Analyze total investment income and tax liability.
    
    Parameters:
    - trades: List of (sale_price, cost_basis, holding_days)
    - dividends: Dict with 'qualified' and 'ordinary' amounts
    - interest: Dict with 'taxable' and 'tax_exempt' amounts
    - ordinary_income: Other taxable income (salary, etc.)
    - filing_status: 'single' or 'married'
    """
    results = {
        'capital_gains': {'st_gains': 0, 'lt_gains': 0, 'st_losses': 0, 'lt_losses': 0},
        'dividends': dividends,
        'interest': interest,
        'taxes': {}
    }
    
    # Process trades
    for sale, cost, days in trades:
        gain = sale - cost
        is_lt = days > 365
        
        if gain >= 0:
            if is_lt:
                results['capital_gains']['lt_gains'] += gain
            else:
                results['capital_gains']['st_gains'] += gain
        else:
            if is_lt:
                results['capital_gains']['lt_losses'] += abs(gain)
            else:
                results['capital_gains']['st_losses'] += abs(gain)
    
    # Net capital gains
    cg = results['capital_gains']
    net_st = cg['st_gains'] - cg['st_losses']
    net_lt = cg['lt_gains'] - cg['lt_losses']
    
    # Calculate taxes
    total_ordinary = ordinary_income + dividends.get('ordinary', 0) + interest.get('taxable', 0)
    
    # Add short-term gains to ordinary (if positive)
    if net_st > 0:
        total_ordinary += net_st
    
    ordinary_tax = calculate_ordinary_tax(total_ordinary, filing_status)
    
    # Long-term gains and qualified dividends
    lt_income = max(0, net_lt) + dividends.get('qualified', 0)
    ltcg_rate = get_ltcg_rate(total_ordinary + lt_income, filing_status)
    ltcg_tax = lt_income * ltcg_rate
    
    # Capital loss deduction (up to $3,000)
    total_net_loss = 0
    if net_st < 0 and net_lt < 0:
        total_net_loss = abs(net_st) + abs(net_lt)
    elif net_st < 0:
        total_net_loss = max(0, abs(net_st) - max(0, net_lt))
    elif net_lt < 0:
        total_net_loss = max(0, abs(net_lt) - max(0, net_st))
    
    loss_deduction = min(total_net_loss, 3000)
    carryover = max(0, total_net_loss - 3000)
    
    results['taxes'] = {
        'ordinary_tax': ordinary_tax,
        'ltcg_tax': ltcg_tax,
        'total_tax': ordinary_tax + ltcg_tax,
        'loss_deduction': loss_deduction,
        'loss_carryover': carryover,
        'net_st_gain': net_st,
        'net_lt_gain': net_lt
    }
    
    return results

# Example analysis
print("INVESTMENT INCOME ANALYSIS")
print("="*50)

trades = [
    (15000, 10000, 400),  # LT gain of $5,000
    (8000, 6000, 200),    # ST gain of $2,000
    (4000, 5000, 180),    # ST loss of $1,000
]

dividends = {'qualified': 2000, 'ordinary': 500}
interest = {'taxable': 1000, 'tax_exempt': 500}

analysis = analyze_investment_income(
    trades, dividends, interest, 
    ordinary_income=75000, 
    filing_status='single'
)

print("\nCapital Gains Summary:")
cg = analysis['capital_gains']
print(f"  Short-term gains: ${cg['st_gains']:,.0f}")
print(f"  Short-term losses: ${cg['st_losses']:,.0f}")
print(f"  Long-term gains: ${cg['lt_gains']:,.0f}")
print(f"  Long-term losses: ${cg['lt_losses']:,.0f}")

print("\nDividend Income:")
print(f"  Qualified: ${dividends['qualified']:,.0f}")
print(f"  Ordinary: ${dividends['ordinary']:,.0f}")

print("\nInterest Income:")
print(f"  Taxable: ${interest['taxable']:,.0f}")
print(f"  Tax-exempt: ${interest['tax_exempt']:,.0f}")

print("\nTax Summary:")
taxes = analysis['taxes']
print(f"  Net ST Gain: ${taxes['net_st_gain']:,.0f}")
print(f"  Net LT Gain: ${taxes['net_lt_gain']:,.0f}")
print(f"  Ordinary Income Tax: ${taxes['ordinary_tax']:,.0f}")
print(f"  LTCG/Qualified Div Tax: ${taxes['ltcg_tax']:,.0f}")
print(f"  Total Investment Tax: ${taxes['total_tax']:,.0f}")
````

## Summary

### Key Takeaways

1. **Taxes significantly impact returns** - A 30% tax rate turns 10% return into 7%

2. **Three types of investment income**:
   - Capital gains (from selling assets)
   - Dividends (company profit distributions)
   - Interest (from bonds, savings)

3. **Realized vs unrealized**:
   - Unrealized = paper gains, no tax
   - Realized = sold, tax is due

4. **Holding period matters**:
   - Short-term (≤1 year): Ordinary rates (10-37%)
   - Long-term (>1 year): Preferential rates (0-20%)

5. **Tax deferral compounds wealth** - Delaying taxes allows full return to compound

### Key Formulas

```
Capital Gain = Sale Price - Cost Basis

After-Tax Return = Pre-Tax Return × (1 - Tax Rate)

Tax Deferral Benefit increases with time!
```
